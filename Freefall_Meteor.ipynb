Sebuah meteor jatuh ke Bumi dari ketinggian awal 10.000 meter. Tentukan ketinggian meteor pada detik ke-5 dengan menggunakan pendekatan analitik dan numerikal. Hitung juga waktu yang dibutuhkan sejak meteor dilepaskan hingga mencapai permukaan Bumi.


> Diketahui
* Ketinggian awal meteor (h0): 10.000 meter
* Percepatan gravitasi (g): 9.81 m/s²
* Waktu awal 	: 0s
* Delta t		: 1s

> Ditanyakan
* Ketinggian meteor pada detik ke 5 dengan menggunakan pendekatan analitik.
*	Ketinggian meteor pada detik ke 5 dengan menggunakan pendekatan numerik.
*	Total waktu yang dibutuhkan sejak meteor dilepaskan hingga mencapai permukaan Bumi dengan pendekatan analitik
* Total waktu yang dibutuhkan sejak meteor dilepaskan hingga mencapai permukaan Bumi dengan pendekatan numerik

> Analisis
*	Berikan analisis perbandingan total waktu pendekatan analitik dan numerikal
*	Berikan analisis perbandingan hasil posisi meteor pada detik ke-5 pendekatan analitik dan numerikal


```python
# improt library
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Data inputan

# input
y = 10000     # posisi awal (m)
v = 0         # kecepatan awal (m/s)
a = -9.8      # percepatan --> gravitasi (m/s^2)
t = 0         # waktu awal (s)
dt = 1        # delta t (s)

# inisialisasi array
y_arr = [y]
v_arr = [v]
t_arr = [t]
```

```python
# Solusi numerikal

# update
while y >= 0:
    v += a * dt      # update kecepatan
    y += v * dt      # update posisi
    t += dt          # update waktu
    if y < 0:
        break
    # simpan output masing masing variabel
    v_arr.append(v)
    y_arr.append(y)
    t_arr.append(t)
```

```python
# Solusi Analitik

# exact solution ⟹ y(t) = y(0) + (1/2gt^2)
y_arr_ex = []
for t_val in t_arr:
    y_ex = y_arr[0] + (0.5 * a * t_val**2)
    y_arr_ex.append(y_ex)

# total time ⟹ t_tot = sqrt((2*y(0)) / g)
t_ex = np.sqrt((2 * -y_arr[0])/a)
```

# Menghitung total waktu meteor sampai ke permukaan bumi

```python
# Visualisasi posisi

# Plot output numerical solution
plt.plot(t_arr, y_arr, c='b', label='numerical')

# Plot output analytical solution
plt.plot(t_arr, y_arr_ex, c='r', label='analytical')

# plot properties
plt.xlabel('time')
plt.ylabel('position')
plt.xlim(-0.5,50)
plt.ylim(-5,10000)
plt.legend()
plt.show()
```

```python
# Bandingkan total waktu
print('Numerical vs Analytical')
print('Total time (s): {:.2f} vs {:.2f}'.format(t_arr[-1], t_ex))
```

Dari hasil komputasi di atas, terlihat bahwa selisih total waktu solusi analitik dengan solusi numerik adalah 1,18 detik.

```python
# Calculate gap/error
error_arr = np.array(y_arr) - np.array(y_arr_ex)
max_error = np.max(np.abs(error_arr))
avg_error = np.mean(np.abs(error_arr))

print('\nGap/Error Analysis:')
print('Max Error (m):', max_error)
print('Average Error (m):', avg_error)
```

# Menghitung posisi meteor pada detik ke- 5

```python
# Visualisasi posisi meteor pada detik ke-5

# plot properties
plt.xlabel('time')
plt.ylabel('position')
plt.xlim(0, 10)             # memperbesar skala sumbu x hingga 10 detik
plt.ylim(9500, 10000)       # memperbesar skala sumbu y hingga 9500-10000 meter

# mark position at t = 5 seconds
index_t5 = t_arr.index(5)
y_t5_numerical = y_arr[index_t5]
y_t5_analytical = y_arr_ex[index_t5]
plt.scatter(5, y_t5_numerical, c='b', label='numerical (t=5)')
plt.scatter(5, y_t5_analytical, c='r', label='analytical (t=5)')
plt.text(5, y_t5_numerical, f'({int(y_t5_numerical)})', fontsize=8, ha='right', va='bottom')
plt.text(5, y_t5_analytical, f'({int(y_t5_analytical)})', fontsize=8, ha='right', va='bottom')

# show legend
plt.legend()

plt.show()

print(f"Pada detik ke-5, posisi meteor dengan pendekatan numerikal adalah {y_t5_numerical} meter")
print(f"Pada detik ke-5, posisi meteor dengan pendekatan analitik adalah {y_t5_analytical} meter")
```

Dari hasil komputasi di atas, terlihat bahwa terdapat selisih antara solusi analitik dengan solusi numerik sebesar 0.24 meter

# Kesimpulan


> Pada analisis kesalahan (error analysis), kita melihat bahwa ada perbedaan antara hasil numerik dan analitik, dengan gap **error maksimum sebesar 215.6 meter** dan **error rata-rata sebesar 107.8 meter**. Bahkan, pada saat menghitung posisi meteor pada detik ke-5, perbedaan antara solusi analitik dan solusi numerik hanya sebesar 0,24 meter.
